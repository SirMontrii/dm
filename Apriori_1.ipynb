```python
from efficient_apriori import apriori
import pandas as pd
```

```python
data=pd.read_csv("groceriesDataset(1).csv")
df=pd.DataFrame(data)
df
```

```python
df.fillna(0,inplace=True)
```

```python
transactions = []
for index, row in df.iterrows():
    # Filter out NaN values and convert items to strings
    transaction = [str(item) for item in row if pd.notna(item) and item!=0 and str(item).isnumeric()==False]
    transactions.append(transaction)
```

```python
transactions
```

```python
itemsets_1, rules_1 = apriori(transactions, min_support=0.04, min_confidence=0.15)
```

```python
itemsets_1
```

```python
rules_1
```

```python
itemsets_2, rules_2 = apriori(transactions, min_support=0.6, min_confidence=0.6)
```

```python
itemsets_2
```

```python
rules_2
```
